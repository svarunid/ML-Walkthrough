## Workflow

```python
# 0. Import Matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Data preparation
x = [1,2,3,4,5,6,7]
y = [33,66,33,99,33,66,33]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot Data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",xlabel="X-Axis",ylabel="Y-Axis")

# 5. Save and show the figure
fig.savefig("sample-plot.png")
```

## Figures with NumPy Arrays

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Line plot
fig, ax = plt.subplots()
ax.plot(x,x**2);
```

```python
# Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Bar plot - Creating a plot using dictionary
nut_butter_price = {
    "Alomnd-Butter": 10,
    "Cashew-Butter": 12,
    "Peanut-Butter": 8
}

fig, ax = plt.subplots()
ax.bar(nut_butter_price.keys(),nut_butter_price.values())
ax.set(title="Nut Butter Store",ylabel="Prices ($)");
```

```python
# Horizontal Bar plot
fig, ax = plt.subplots()
ax.barh(list(nut_butter_price.keys()),list(nut_butter_price.values()))
ax.set(title="Nut Butter Store",xlabel="Prices ($)");
```

```python
# Histogram
fig, ax = plt.subplots()
ax.hist(np.random.randn(1000));
```

```python
# Subplots - Method 1
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,10))
ax1.plot(x,x**2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_price.keys(),nut_butter_price.values()),
ax4.hist(np.random.randn(1000));
```

```python
# Subplots - Method 2
fig, ax = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,10))
ax[0,0].plot(x,x**2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_price.keys(),nut_butter_price.values()),
ax[1,1].hist(np.random.randn(1000));
```

## Making Plots from Pandas DataFrame

```python
# import our dataframe
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]","")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2].astype(int)
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2021",periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum() # Cumulative Sum
car_sales
```

```python
# line Plot
car_sales.plot(x="Sale Date",y="Total Sales");
```

```python
# Scatter Plot - Method 1
car_sales.plot(x="Price",y="Odometer (KM)",kind="scatter");
```

```python
# Bar Plot - Method 2
car_sales.plot.bar(x="Make", y="Odometer (KM)");
```

```python
# Histogram - Method 2
car_sales["Odometer (KM)"].plot.hist();
```

## Heart-Disease Analysis

Relpicate the below image

<img src="matplotlib-anatomy-of-a-plot.png">

```python
heat_disease  = pd.read_csv("heart-disease.csv")
heat_disease
```

```python
# Filter the age(>50)
over_50 = heat_disease[heat_disease["age"] > 50]
over_50
```

```python
# Define the plot style
plt.style.use("seaborn-whitegrid")

# Create a fig and 2 axes
fig, (ax1, ax2) = plt.subplots(nrows=2,
                               figsize=(10,10),
                               sharex=True)

# Add figure title
fig.suptitle("Heart Disease analysis",
             fontsize=16,
             fontweight="bold")

# Add data to the axes
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")
ax2.scatter(x=over_50["age"],
            y=over_50["thalach"],
            c=over_50["target"],
            cmap="winter")

# Add titles to the axes
ax1.set(ylabel="Cholestrol",
        title="Cholestrol vs Age")
ax2.set(xlabel="Age",
        ylabel="Max Heart Rate",
        title="Max Heart Rate vs Age")

# Set axess limits
ax1.set_xlim(50,80)
ax2.set_ylim(60,200)

# Set axes legends
ax1.legend(*scatter.legend_elements(),
           title="Target")
ax2.legend(*scatter.legend_elements(),
           title="Target")

# Horizontal line in axes
ax1.axhline(y=over_50["chol"].mean(),
            c="r",
            linestyle="--",);
ax2.axhline(y=over_50["thalach"].mean(),
            c="r",
            linestyle="--",);
```
